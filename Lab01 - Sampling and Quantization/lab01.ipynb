```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("relogio.tif", cv2.IMREAD_GRAYSCALE)

print('Tamanho da imagem : ', img.shape)

linhas = img.shape[0]
colunas = img.shape[1]

nova_img1 = np.zeros((linhas // 4, colunas // 4))
nova_img2 = np.zeros((linhas // 8, colunas // 8))
nova_img3 = np.zeros((linhas // 17, colunas // 17))

for i in range(2, colunas - 2, 4):
    for j in range(2, linhas - 2, 4):
        bloco = img[j-2:j+3, i-2:i+3] 
        nova_img1[j // 4, i // 4] = np.mean(bloco)

plt.figure()
plt.imshow(nova_img1, cmap='gray', vmin=0, vmax=255)
plt.title("300 DPI")

for i in range(4, colunas - 4, 8):
    for j in range(4, linhas - 4, 8):
        bloco = img[j-4:j+5, i-4:i+5] 
        nova_img2[j // 8, i // 8] = np.mean(bloco)

plt.figure()
plt.imshow(nova_img2, cmap='gray', vmin=0, vmax=255)
plt.title("150 DPI")

for i in range(8, colunas - 17, 17):
    for j in range(8, linhas - 17, 17):
        bloco = img[j-8:j+9, i-8:i+9] 
        nova_img3[j // 17, i // 17] = np.mean(bloco)

plt.figure()
plt.imshow(nova_img3, cmap='gray', vmin=0, vmax=255)
plt.title("72 DPI")

plt.show()
```

2)	Utilizando a imagem ctskull-256.tif, refaça a quantização de seus pixels utilizando de 7 a 1 bit(s) por pixel.  Fazer os cálculos por meio do seu algortimo. Não usar o opencv para a conversão

```python

img2 = cv2.imread("ctskull-256.tif", cv2.IMREAD_GRAYSCALE)

print('Tamanho da imagem : ', img2.shape)

linhas = img2.shape[0]
colunas = img2.shape[1]

for i in range(colunas):
    for j in range(linhas):
        if img2[j, i] < 32:
            img2[j, i] = 0
        if img2[j, i] < 64 and img2[j, i] >= 32:
            img2[j, i] = 1
        if img2[j, i] < 96 and img2[j, i] >=64 :
            img2[j, i] = 2
        if img2[j, i] < 128 and img2[j, i] >= 96:
            img2[j, i] = 3
        if img2[j, i] < 160 and img2[j, i] >= 128:
            img2[j, i] = 4
        if img2[j, i] < 192 and img2[j, i] >= 160:
            img2[j, i] = 5
        if img2[j, i] < 224 and img2[j, i] >=192:
            img2[j, i] = 6
        if img2[j, i] < 256 and img2[j, i] >= 224:
            img2[j, i] = 7

plt.imshow(img2, cmap='gray');
```
